```python
import urllib.request as res 
import requests
import pandas as pd
from bs4 import BeautifulSoup
import bs4
import csv
import jieba.analyse
csvfile="PythonUrl.csv"
for i in range(1):
    url="https://technews.tw/category/mobiledevice/page/"+str(i+1)+"/"
    req=res.Request(url, headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"})
    with res.urlopen(req) as response:
        data=response.read().decode("utf-8")#read支援html、CSS、JS
        #用utf-8解碼 中文資料才能夠比較好的被處理
    root=bs4.BeautifulSoup(data, "html.parser")
    titles=root.find_all("h1", class_="entry-title")
    with open(csvfile,"w",newline='',encoding="utf-8") as fp:
        writer=csv.writer(fp)
        for title in titles:
             titleUrl=[]
             titleUrl.append(title.a.get('href'))
             writer.writerow(titleUrl)
```
