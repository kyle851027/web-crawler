```python
import urllib.request as res 
import requests
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import csv
import jieba
import jieba.posseg as pseg
import sklearn
import jieba.analyse
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
with open('PythonUrl2.csv', newline='') as csvfile:#打開PythonUrl.csv
        rows = csv.reader(csvfile)#讀檔
        for row in rows:#建立迴圈 依次抓取網址出來
            req=res.Request(row[0], headers={"user-agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36"})
            with res.urlopen(req) as response:#打開網站
                data=response.read().decode("utf-8")#抓取資訊
            root=bs4.BeautifulSoup(data,"lxml")#將data丟給beautifulsoup 它會用lxml幫我們解析爬下來的網頁資訊
            contents=root.find_all("div", class_="article__introduction")#find:這次只需要找一筆資料 就是新聞內容
        #從網頁資訊尋找標籤 要尋找的標籤為div 篩選條件為 class=indent
            newstitle=root.find("title")#抓取新聞標題
            for content in contents:#建立迴圈
                seg_list=jieba.cut(content.get_text(), cut_all=False)#jueba.cut()函數 分割句子成有意義的字詞元素
                                                                    #cut_all=False:精確模式，將句子最精確地切開，以便文本分析
                contentList_1=(" ".join(seg_list))#join()函數:序列中的元素以指定的字詞連接生成一個新的字詞串 每增加一個字詞元素就加一個空白
                contentList=[]#建立一個空群集待會將資料放入
                contentList.append(contentList_1)#append()函數:將資料放入群集
                vectorizer=CountVectorizer()#該函數會將文本中的詞語轉換為詞頻矩陣，矩陣元素a[i][j] 表示j詞在i類文本下的詞頻
                transformer=TfidfTransformer()#該函數會統計每個詞語的tf-idf權值
                tfidf=transformer.fit_transform(vectorizer.fit_transform(contentList))
                     #第一個fit_transform是計算tf-idf，第二個fit_transform是將contentList轉為詞頻矩陣
                word=vectorizer.get_feature_names()#獲取詞袋模型中的所有字詞
                weight=tfidf.toarray()#將tf-idf矩陣抽取出來利用.toarray()函數將轉成陣列weight，元素a[i][j]表示j詞在i類文本中的tf-idf權重
                for i in range(len(weight)):#打印每類文本的tf-idf詞語權重，第一個for遍歷所有文本，
                                            #len()函數:抓出weight陣列長度得知文本數量決定迴圈大小
                    print("新聞標題:",newstitle.string)
                                            #.string:將新聞標題的文字印出來 不加string的話會有其他標籤
                    print("關鍵字:")
                    for j in range(len(word)):#第二個for抓出某一類文本下的詞語權重
                                            #len()函數:抓出一文本內字詞數量得知決定迴圈大小
                       print (word[j],weight[i][j])
                  
               
                
```
